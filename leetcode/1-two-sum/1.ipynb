```python
import seaborn as sns
from nltk.tokenize import word_tokenize
from collections import Counter
import nltk

nltk.download('punkt')

# load dataset
df = sns.load_dataset("tips")

# convert to text
docs = df[['day','time','sex']].astype(str).agg(' '.join, axis=1)

# join all rows into one text
text = " ".join(docs)

# tokenization
tokens = word_tokenize(text)

# frequency
freq = Counter(tokens)

# top 5 common words
for word, count in freq.most_common(5):
    print(word, ":", count)
```

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = "Elon Musk is the CEO of SpaceX and Tesla," \
        " and he was born in South Africa." \

tokens = word_tokenize(text)
tags = pos_tag(tokens)
tree = ne_chunk(tags)

print(tree)
```
