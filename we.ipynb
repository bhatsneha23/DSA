```python
import random

text = "the cat sat on the mat the cat ate the rat"
words = text.split()

# missing step - build model
model = {}
for i in range(len(words) - 1):
    model.setdefault(words[i], []).append(words[i+1])

# your code
word = "the"
result = [word]
for _ in range(14):
    word = random.choice(model.get(word, words))
    result.append(word)

print("15 words:", " ".join(result))
```
